```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import urllib.request
def url_to_image(url):
  resp = urllib.request.urlopen(url)
  image = np.asarray(bytearray(resp.read()), dtype='uint8')
  image = cv2.imdecode(image, cv2.IMREAD_COLOR)

  return image

img = url_to_image('https://upload.wikimedia.org/wikipedia/commons/5/50/Vd-Orig.png')
kernel = np.array([[1,3,1],[3,6,3,],[1,3,1]])*(1/24)

output2=cv2.GaussianBlur(img,(3,3),0)
output3=cv2.GaussianBlur(img,(7,7),0)
merge=np.hstack((output2,output3))
cv2_imshow(merge)
```
